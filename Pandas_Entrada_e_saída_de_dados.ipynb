# Incluindo ID aos alunos

```python
import numpy as np
np.random.seed(123) #Colocar um seed no random significa que vai ser gerado um n√∫mero aleat√≥rio depois do n√∫mero dado como par√¢metro, no caso '123', e toda vez que rodar ter√° a mesma sequ√™ncia
```

```python
len(nomes)
```

```python
nomes['id_aluno'] = np.random.permutation(len(nomes)) + 1
nomes.head(5)
```

```python
nomes['dominio'] = np.random.choice(dominios, len(nomes)) #choice() escolhe uma das op√ß√µes de forma aleat√≥ria
nomes
```

```python
nomes['email'] = nomes['nome'].str.cat(nomes['dominio']).str.lower() #Para usar a fun√ß√£o .cat() √© necess√°rio atribuir tipo
nomes.head(5)
```

# Atribuindo alunos aos cursos

```python
#Abaixo estaremos atribuindo matriculas para a tabela.
#np.ceil faz n√£o aparecer matricula 0
nomes['matriculas'] = np.ceil(np.random.exponential(size=len(nomes)) * 1.5).astype(int)
nomes.head()
```

```python
nomes.matriculas.describe()
```

Atribuindo quais cursos os alunos est√£o matriculados

```python
nomes.head()
```

```python
todas_matriculas = []
x = np.random.rand(20)
prob = x/sum(x)
```

```python
x
```

```python
for index, row in nomes.iterrows():
  id = row.id_aluno
  print(id)
  matriculas = row.matriculas
  print(matriculas)
  for i in range(matriculas):
    mat = [id, np.random.choice(cursos.index, p = prob)]
    print(mat)
    todas_matriculas.append(mat)
```

```python
todas_matriculas
```

```python
matriculas = pd.DataFrame(todas_matriculas, columns = ['id_aluno', 'id_curso'])
matriculas.head()
```

```python
nomes.head()
```

```python
matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns={'id_aluno': 'quantidade_de_alunos'})
```

# Sa√≠da em diferentes formatos

```python
matriculas_por_curso = matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns={'id_aluno': 'quantidade_de_alunos'})
```

```python
matriculas_por_curso.head()
```
